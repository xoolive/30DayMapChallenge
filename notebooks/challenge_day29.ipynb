```python
import cartopy.io.shapereader as shpreader
from geopandas import GeoDataFrame


countries_110 = GeoDataFrame.from_file(
    shpreader.natural_earth(
        resolution="110m",
        category="cultural",
        name="admin_0_countries",
    )
)
```

```python
countries = list(
    c.strip()
    for c in """
    United States of America
    Bahamas
    Cayman Islands
    Liberia
    Palau
    The Federated States of Micronesia
    Marshall Islands
""".split(
        "\n"
    )
)
countries
```

```python
c = countries_110.query("ADMIN not in @countries").assign(t=100)
```

```python
import altair as alt

base = (
    alt.Chart(c)
    .mark_geoshape(stroke="#fff")
    .encode(
        alt.Color(
            "t",
            legend=alt.Legend(title="Temperature in °C", orient="bottom"),
        ),
        alt.Tooltip("ADMIN"),
    )
    .project("equalEarth")
    .properties(width=500, height=300, title="Temperature of boiling water in °C")
    .configure_view(stroke=None)
    .configure_legend(
        orient="bottom",
        titleFontSize=14,
        labelFontSize=13,
        titleFont="Fira Sans",
        labelFont="Ubuntu",
    )
    .configure_title(anchor="start", font="Fira Sans", fontSize=18, dy=-10)
)

base
```

```python
base.save("../contributions/challenge_day29.json")
base.save("../contributions/challenge_day29.svg")
```
