```python
from cartes.crs import Amersfoort, PlateCarree
from cartes.utils.features import countries
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText

fig, ax = plt.subplots(
    figsize=(10, 10),
    dpi=300,
    subplot_kw=dict(projection=Amersfoort()),
)

bb1.plot(ax, color="#ff9b00")
bb2.plot(ax, color="#211e1e")
boundary.query('type_ == "way"').plot(
    ax, edgecolor="#bab0ac", linestyle="dashed", linewidth=0.7, facecolor="none"
)

ax.set_extent((4.85, 5, 51.39, 51.47))
ax.spines["geo"].set_visible(False)

text = AnchoredText(
    "Baarle-Nassau (NL)",
    loc=1,
    frameon=False,
    prop={"size": 24, "fontname": "Fira Sans", "color": "#ff9b00"},
)
ax.add_artist(text)

text = AnchoredText(
    "Baarle-Hertog (B)",
    loc=3,
    frameon=False,
    prop={"size": 24, "fontname": "Fira Sans", "color": "#211e1e"},
)
ax.add_artist(text)


fig.set_tight_layout(True)
fig.savefig("../contributions/challenge_day22.png", dpi=300, background="white")
```
